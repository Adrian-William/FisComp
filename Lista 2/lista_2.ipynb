```python
def bhaskara(a, b, c):
    '''
     Calcula bhaskara da função " a*X^2 + b*X +c = 0 " e retorna os possiveis valores de X.

    Argumentos
    ----------
    a : float
        Constande que acompanha X^2.
    b : float
        Constante que acompanha X.
    c : float
        Constante.

    Retorna
    -------
    Tupla
        O possiveis valores de X.
    '''


    delta = (b**2-4*a*c)**(1/2)
    x = [(-b+delta)/(2*a),(-b-delta)/(2*a)]

    return x
```

```python
import numpy as np
import matplotlib.pyplot as plt

#calculando valores de x e y
x=np.linspace(-5,5,1000)
a,b,c = (4,2,-10)
y= a*x**2+b*x+c

#calculando raizes x
X = np.array(bhaskara(a, b, c))
Y = a*(X)**2+b*(X)+c

#plotando grafico
plt.figure()
plt.plot(x, y)
plt.grid()
plt.scatter(X,Y,color='black')
```

```python
import numpy as np
import matplotlib.pyplot as plt

#calculando valores de x e y
x=np.linspace(-20,20,1000)
a,b,c=2,0,4
y= a*np.exp(-((x-b)/c)**2)

#salvando valores de  x e y
data=(x,y)
np.savetxt('data.txt',data)
```

```python
import numpy as np
import matplotlib.pyplot as plt

#carregando valores de x e y
x,y = np.loadtxt('data.txt')

#plotando grafico
plt.figure()
plt.plot(x, y)
plt.grid()
```

```python
import numpy as np
import matplotlib.pyplot as plt

#calculando valores de x e y
x=np.linspace(-5,5,1000)
a,b,c = (4,2,-10)
y= a*x**2+b*x+c
#calculando raizes de x
X = np.array(bhaskara(a, b, c))
Y = a*(X)**2+b*(X)+c

data,raizes=[x,y],[X,Y]

#salvando valores de  x e y
np.savetxt('data3.txt',np.ascontiguousarray(data))

#salvando raizes de x
np.savetxt('raizes3.txt',np.ascontiguousarray(raizes))

```

# Atividade avançada

Faça um gráfico das curvas paramétricas:

    Deltoid:

      x = 2 cos ⁡ θ + cos ⁡ 2 θ
      y = 2 sin ⁡ θ − sin ⁡ 2 θ

    Espiral de Galileu:
      r = θ^2

    Fey’s function:
      r = e cos ⁡ θ − 2 cos ⁡ 4 θ + sin 5 ⁡ ( θ / 12 )


```python
import numpy as np
import matplotlib.pyplot as plt

#calculando valores da equação
teta = np.linspace(0,2*np.pi,1000)

#tranformando para cordenadas cartezianas
x = 2*np.cos(teta)+np.cos(2*teta)
y = 2*np.sin(teta)-np.sin(2*teta)

#plotando grafico
plt.figure()
plt.plot(x, y)
plt.grid()
```
